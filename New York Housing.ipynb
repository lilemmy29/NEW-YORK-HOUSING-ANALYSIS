```python
import pandas as pd 
import plotly.express as px
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import r2_score,mean_squared_error
import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\user\Desktop\NY Housing Market\NY-House-Dataset.csv")
```

```python
df.drop_duplicates(inplace = True, ignore_index=True)
```

```python
df['BATH'] = df.BATH.astype(int)
```

```python
df['PROPERTYSQFT'] = df.PROPERTYSQFT.astype(int)
```

```python
df.columns
```

```python
df.drop(columns=['V'],inplace=True)
```

```python
df.SUBLOCALITY.nunique()
```

```python
X = df[['TYPE','BEDS','BATH','PROPERTYSQFT','STATE','SUBLOCALITY', 'LOCALITY']]
y= df['PRICE']
x = pd.get_dummies(X,drop_first=True).astype(int)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=45)
```

```python
x_train
```

```python
rf_model = RandomForestRegressor().fit(x_train,y_train)
```

```python
rf_model.predict(x_test).tolist()
```

```python
r2_score(y,rf_model.predict(x))
```

```python
mean_squared_error(y,rf_model.predict(x))
```

```python
ols_model = sm.OLS(y,sm.add_constant(x))
```
